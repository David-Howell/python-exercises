* How many different combinations are there of 2 letters from "abcd"?

```python
two_letter_combos = it.combinations('abcd', 2)
combs = [''.join(i) for i in two_letter_combos]
print(combs)
c = 0
for i in combs: c += 1
print(c)
```

* How many different permutations are there of 2 letters from "abcd"?

```python
two_letter_permutations = it.permutations('abcd', 2)
perms = [''.join(i) for i in two_letter_permutations]
print(perms)
c = 0
for i in perms: c += 1
print(c)
```

* Grand total of balances for all users

```python
def norm_bal(a):
    return float(a.strip('$').replace(',',''))
```

```python
def total_balance(x):
    totaly_balances = 0

    for l in x:
        totaly_balances += norm_bal(l['balance'])
#         totaly_balances += float(((l['balance']).replace('$','')).replace(',',''))
    return totaly_balances
total_balance(profiles)
```
